```python
import pandas as pd
import joblib
```

```python
df = pd.read_csv('newspapers_ger_1914_part_1_only_ads_REGIONS.csv')
```

```python
df
```

```python
df = df.dropna(subset=['text'])
```

```python
# Regions classification

svm_model = joblib.load('job_ads_identification_model.joblib')
vectorizer = joblib.load('tfidf_vectorizer.joblib')

def predict_region(text, model, vectorizer):
    X_test_tfidf = vectorizer.transform(text)
    y_pred = model.predict(X_test_tfidf)
    return y_pred

df['classification'] = predict_region(df['text'], svm_model, vectorizer)
```

```python
df.classification
```

```python
classified_ads = df[df['classification'] == 1]
print(classified_ads['text'])
```

```python
classified_ads.to_csv('classified_ads.csv', index=False)
```
